```python
# DEPENDÊNCIA E INDEPENDEÊNCIA

"""
    Em regra, dizemos que 2 eventos (E e F) são dependentes quando
    sabemos algo sobre a ocorrência de E que nos dá informações sobre a ocorrência de F (e vice-versa).
    Sem essa relação, os eventos são independetes.
    
    Exemplo (Eventos independentes):
        Uma moeda é jogada 2 vezes;
        Se o primeiro resultado for 'cara', (Evento 1)
        Esta informação não indica qual será o segundo resultado. (Evento 2)
    
    Exemplo (Eventos dependentes):
        Uma moeda é jogada 2 vezes;
        Se alguém tiver dito que os 2 resultados seriam 'coroa' (Evento 1)
        Mas, o primeiro resultado foi 'cara', (Evento 2)
        Este evento diz algo sobre o evento anterior.
    
    Dois eventos são independentes se a probabilidade de ambos ocorrerem for igual ao produto
    da probabilidade de cada um ocorrer:
    
    P(E, F) = P(E) . P(F)
    """
```

```python
# PROBABILIDADE CONDICIONAL

"""
    Se, dois eventos (E e F) não forem independentes:
        Se a probabilidade de F ocorrer não for 0:
            A probabilidade de E acontecer, se F acontecer, é:
                
                P(E|F) = P(E, F) / P(F)
        
        Se a probabilidade de E ocorrer não for 0:
            A probabilidade de F acontecer, se E acontecer, é:
            
                P(F|E) = P(E, F) / P(E)
    """

import enum, random

"""
    PERGUNTA: 
    Um casal terá 2 filhos. 
    A probabilidade de nascerem 2 filhos é '1/4',
    A probabilidade de nascerem 1 filho e uma 1 filha é '1/2',
    A probabilidade de nascerem 2 filhas é '1/4',
    Agora...
    Dado que 1 primeira é uma menina, qual a probabilidade deles terem 2 meninas?
"""

class Kid(enum.Enum):   # Enum é um conjunto tipado de valores enumerados que deixa o código mais descritivo e legível.
    BOY = 0
    GIRL = 1

def random_kid() -> Kid:
    return random.choice([Kid.BOY, Kid.GIRL])

both_girl = 0
older_girl = 0
either_girl = 0

random.seed(0)

for _ in range(10000):
    younger = random_kid()      # filho 1 (mais novo)
    older = random_kid()        # filho 2 (mais velho)
    if older == Kid.GIRL:
        older_girl += 1
    if older == Kid.GIRL and younger == Kid.GIRL:
        both_girl += 1
    if older == Kid.GIRL or younger == Kid.GIRL:
        either_girl += 1

print("P (both | older): ", both_girl / older_girl)         # 50 % de serem 2 filhas, dado que a primeira é filha
print("P (both | either): ", both_girl / either_girl)       # 33 % de 2 filhas, dado que uma das duas é filha
```

```python
# O TEOREMA DE BAYES

"""
    Imagina que precisamos saber a probabilidade do evento E condicionado à ocorrência do evento F
    Porém, só temos a informação sobre a probabilidade da ocorrência de F condicionado a E.
    
    P(E|F) = P(E, F) / P(F) = P(F|E) . P(E) / P(F)
    
    É possível dividir o evento F em 2 eventos mutuamente exclusivos "F e E" e "F e não E"
    
    P(F) = P(F, E) + P(F, ~E)
    
    Logo:
    
    P(E|F) = P(F|E) . P(E) / [P(F|E) . P(E) + P(F|~E) . P(~E)]
    """
```

```python
# VARIÁVEIS ALEATÓRIAS

"""
    Uma variável aleatória é aquela cujos valores possíveis estão associados a uma distribuição
    de probabilidade.
    
    As vezes, menciona-se o valor esperado da variável aleatória, a média dos seus valores ponderados,
    pela sua probabilidade.
    Isto é, no caso de uma variável aleatória igual a 1 ou a 0...
    Cada valor possui P = 0.5
    bem como, o valor esperado é (0 . 0.5 + 1 . 0.5) = 0.5
    """
```

```python
# DISTRIBUIÇÕES CONTÍNUAS

"""
    A ditribuição uniforme atribui um peso igual a todos os números entre 0 e 1.
    Ela funciona como uma função de densidade de probabilidade,
    pois, assim, a probabilidade de identificar um valor em um determinado intervalo
    é igual à integral da função de densidade sobre o intervalo.
    
    No python, o random.random é uma variável (pseudo-)aleatória com uma densidade uniforme.
    """

def uniform_pdf(x: float) -> float:
    return 1 if 0 <= x  < 1 else 0

def uniform_cdf(x: float) -> float:
    """Retorna a probabilidade de uma variável aleatória uniforme ser <= x"""
    if x < 0:
        return 0
    elif x < 1:
        return x
    else:
        return 1
```

```python
# A DISTRIBUIÇÃO NORMAL

"""
    A distribuição normal se expressa na clássica curva em forma de sino e é totalmente determinada
    por dois parâmetros: sua média e seu desvio-padrão    
    f (x | mu, sigma)
    """

import math
import matplotlib.pyplot as plt

SQRT_TWO_PI = math.sqrt(2 * math.pi)

def normal_pdf(             # PDF = Probability Density function
        x: float,           # valores / funções da distribuição
        mu: float = 0,      # média
        sigma: float = 1    # desvio padrão
        ) -> float:
    """Fornece a distribuição normal de um conjunto de valores"""
    return (math.exp(-(x-mu) ** 2 / 2 / sigma ** 2) / (SQRT_TWO_PI * sigma))

xs = [x/10.0 for x in range(-50, 50)]

plt.plot(
    xs, 
    [normal_pdf(x, sigma = 1) for x in xs], 
    '-', 
    label='mu = 0, sigma = 1'
)
plt.plot(
    xs,
    [normal_pdf(x, sigma = 2) for x in xs],
    '--',
    label = 'mu = 0, sigma = 2'
)
plt.plot(
    xs,
    [normal_pdf(x, sigma = 0.5) for x in xs],
    ':',
    label = 'mu = 0, sigma = 0.5'
)
plt.plot(
    xs,
    [normal_pdf(x, mu = -1) for x in xs],
    '-.',
    label = 'mu = -1, sigma = 1'
)

plt.legend()

plt.title("Various Normal pdfs")

plt.show()

# http://en.wikipedia.org/wiki/Error_function

"""for a random variable Y that is normally distributed with mean 0 and standard deviation 1/√2, 
erf x is the probability that Y falls in the range [x,x]."""

def normal_cdf(     # CDF = Cumulative Density function
        x,          # valores / funções da distribuição
        mu=0,       # média
        sigma=1     # desvio padrão
        ):
    return (1+math.erf((x - mu)/math.sqrt(2)/sigma))/2

xs = [x/10.0 for x in range(-50,50)]

plt.plot(
    xs,
    [normal_cdf(x, sigma=1) for x in xs],
    '-',
    label = 'mu = 0, sigma = 1'
)
plt.plot(
    xs,
    [normal_cdf(x, sigma=2) for x in xs],
    '--',
    label = 'mu = 0, sigma = 2'
)
plt.plot(
    xs,
    [normal_cdf(x, sigma=0.5) for x in xs],
    '--',
    label = 'mu = 0, sigma = 0.5'
)
plt.plot(
    xs,
    [normal_cdf(x, mu = -1) for x in xs],
    '-.',
    label = 'mu = -1, sigma = 1'
)

plt.legend(loc = 4)

plt.title("Diversas funções de densidade de distribuição cumulativa")

plt.show()

# http://en.wikipedia.org/wiki/Binary_search_algorithm

"""Ocasionalmente, inverte-se a normal_cdf para se obter a probabilidade correspondente"""

def inverse_normal_cdf(         # divide os intervalos por 2 até encontrar um valor suficientemente próximo do desejado.
        p, 
        mu=0, 
        sigma=1, 
        tolerance=.00001
        ):
    """
    Inverte o CDF para encontrar a probabilidade.
    A função divide em 2 intervalores repetidamente, até diminuir 
        para um Z próximo o suficiente da probabilidade desejada
    Encontra o inverso mais próximo usano a busca binária
        """

    # se não for padrão, computa o padrão e redimensiona
    if mu != 0 or sigma != 1:
        return mu + sigma * inverse_normal_cdf(p, tolerance=tolerance)
    
    low_z = -10.0     # normal_cdf(-10) está muito perto de 0
    hi_z = 10.0       # norma_cdf(10) está muito perto de 1

    while hi_z - low_z > tolerance:
        mid_z = (low_z + hi_z)/2        # considera o ponto do meio e o valor
        mid_p = normal_cdf(mid_z)       # da função de distribuição cumulativa

        if mid_p < p:
            # o ponto do meio ainda está baixo, procura acima
            low_z = mid_z
        elif mid_p > p:
            # o ponto do meio ainda está alto, procura abaixo
            hi_z = mid_z
        else:
            break

    return mid_z
```

```python
# O TEOREMA DO LIMITE CENTRAL

from collections import Counter
import random

"""
    Em essência, uma variável aleatória definida como a média de uma
        grande quantidade de variáveis aleatórias distribuídas independente
        e identicamente é ela mesma aproximadamente distribuída normalmente. 
    
    Uma maneira simples de ilustrar isso é considerando as variáveis 
        aleatórias binomiais, as quais possuem 2 parâmetros n e p.
    Uma variável aleatória Binomial(n, p) é apenas a soma de n variáveis 
        aleatórias independentes Bernoulli(p), e cada uma delas é igual a
        1 com probabilidade p e 0 com probabilidade p-1
    """

def bernoulli_trial(p):
    return 1 if random.random() < p else 0

def binomial(n, p):
    """
    n = quantidade de vezes que a soma será feita
    p = referência para a bernoulli_trial
    """
    return sum(bernoulli_trial(p) for _ in range(n))

def make_hist(p, n, num_points):
    """
    num_points = fator de 'achatamento' do eixo y
    """
    data = [binomial(n, p) for _ in range(num_points)]    # Definimos a série histórica

    # usa um gráfico de barras para exibir as amostrar binomiais atuais
    histogram = Counter(data)    
    
    plt.bar(                                           # função que descreve os valores do gráfico 
        [x - 0.4 for x in histogram.keys()],           # eixo x
        [v/num_points for v in histogram.values()],    # eixo y
        0.8,                                           # largura das barras
        color='0.75'                                   # cor cinza
    )

    mu = p*n                        # cálculo da média
    sigma = math.sqrt(n*p*(1-p))    # cálculo do desvio padrão

    xs = range(min(data), max(data) + 1)              # definição do eixo x
    ys = [                                            # definição do eixo y
        normal_cdf(i + 0.5, mu, sigma) - normal_cdf(i - 0.5, mu, sigma) for i in xs    
    ]

    plt.plot(xs, ys)

    plt.title("Distribuição Binomial vs. Aproximação Normal")

    plt.show()

make_hist(0.75, 100, 10000)
```
