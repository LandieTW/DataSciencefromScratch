```python
# MATPLOTLIB

# http://matplotlib.org/

"""
    seaborn = Fornece uma interface atrativa de desenhos de alto nível para informação gráfica estatística.
    (https://seaborn.pydata.org/tutorial.html)
    (http://bokeh.pydata.org)
"""
# https://seaborn.pydata.org/

"""
    altair = é uma biblioteca python declarativa para visualizações
    (https://altair-viz.github.io/user_guide/data.html)
"""
# https://altair-viz.github.io/

# http://d3js.org

# http://bokeh.pydata.org

"""Usaremos matplotlib.pyplot
    Em essência, o pyplot mantém um estado interno no qual é possível construir uma visualização passo a passo.
    Ao terminar, você pode salvá-la com savefig ou exibí-la com show."""

from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# criar um gráfico de linhas, anos no eixo x, gdp no eixo y
plt.plot(years, gdp, color="green", marker="o", linestyle="solid")

# adicionar um título
plt.title("Nominal GDP")

# adicionar um rótulo ao eixo y
plt.ylabel("Billions of $")

plt.show()
```

```python
# GRÁFICOS DE BARRAS

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# plotar as barras com coordenadas x à esquerda [0, 1, 2, 3, 4], alturas [num_oscars]
plt.bar(range(len(movies)), num_oscars)

# adicionar um título
plt.title("My Favorite Movies")

# rotular o eixo y
plt.ylabel("# of Academy Awards")

# rotular o eixo x com os nomes dos filmes nos centros das barras
plt.xticks(range(len(movies)), movies)

plt.show()
```

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# agrupar as notas por decil, mas coloca o 100 com o 90
histogram = Counter(min(grade//10*10, 90) for grade in grades)

plt.bar([x + 5 for x in histogram.keys()],  # mover as barras para a direita em 5
    histogram.values(),                     # atribuir a altura correta a cada barra
    10,                                     # atribuir a largura 10 a cada barra
    edgecolor=(0,0,0)                       # escurecer as bordas das barras
    )

plt.axis([-5, 105,  # eixo x de -5 a 105
           0, 5])   # eixo y de 0 a 5

plt.xticks([10 * i for i in range(11)]) # rótulos do eixo x em 0, 10, ..., 100

plt.xlabel("Decile")

plt.ylabel("# of Students")

plt.title("Distribution of Exam 1 Grades")

plt.show()
```

```python
# GRÁFICOS DE LINHAS

variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, "g-", label="variance")
plt.plot(xs, bias_squared, "r-", label="bias^2")
plt.plot(xs, total_error, "b:", label="total error")

plt. legend(loc=9)

plt.xlabel("Model complexity")

plt.xticks([])

plt.title("The Bias-Variance Tradeoff")

plt.show()
```

```python
# GRÁFICOS DE DISPERSÃO

friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

# rotular cada ponto
for label, friend_count, minute_count in zip(labels, friends, minutes):
    # colocar o rótulo na respectivo ponto, mas levemente deslocado
    plt.annotate(label,
                 xy = (friend_count, minute_count),
                 xytext = (5, -5),
                 textcoords = 'offset points')
    
plt.title("Daily minutes vs. Number of friends")

plt.xlabel("# of friends")

plt.ylabel("Daily minutes spent on the site")

plt.show()
```
